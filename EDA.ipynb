**Разведывательный анализ**

```python
# Импорт библиотек
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Загрузка датасета
df = pd.read_csv('undefind_для_студентов.csv', sep = ';')
df
```

```python
# Создание стобца "Возраст" на основе столбца "Дата рождения пациента"
df['Дата рождения пациента'] = df['Дата рождения пациента'].apply(lambda x: x.split('-'))
df['Возраст'] = df['Дата рождения пациента'].apply(lambda x: 2024-int(x[0]))
df
```

```python
df.head()
```

```python
df[df['ID истории болезни']=='2e1d0b3f-488a-11ed-ab5a-0050568844e6']
```

```python
df['Флаг нормы кол. показателя'] = df['Флаг нормы кол. показателя'].map(lambda x: 'Норма' if x == 'Норм' else x)
```

**Эритроциты**

```python
df_e = df[(df['Кол. лаб. показатель'] == 'Эритроциты') | (df['Кол. лаб. показатель'] == 'Эритроциты (RBC)') | (df['Кол. лаб. показатель'] == 'RBC')]
df_e
```

```python
g1= sns.violinplot(x=df_e['Пол'], y=df_e['Значение кол. показателя'], data=df_e)
g1.set(ylabel='Содержание эритроцитов, 10**12/л')
g1.set(title='Содержание эритроцитов в крови в зависимости от пола')
```

```python
g2 = sns.boxplot(x=df_e['Пол'], y=df_e['Значение кол. показателя'], orient='v')
g2.set(ylabel='Содержание эритроцитов, 10**12/л')
g2.set(title = 'Содержание эритроцитов в крови в зависимости от пола')
```

```python
g3 = plt.hist(df_e['Возраст'], color = 'blue', edgecolor = 'black',
         bins = int(180/30))
plt.xlabel('Возраст пациентов')
plt.ylabel('Абсолютная частота пациентов')
plt.title('Распределение пациентов по параметру возраст')
```

```python
g4 = sns.barplot(x="Пол", y="Значение кол. показателя", hue="Флаг нормы кол. показателя", data=df_e)
g4.set(ylabel='Содержание эритроцитов, 10**12/л')
g4.set(title = 'Распределение флагов нормы эритроцитов в зависимости от пола')
```
